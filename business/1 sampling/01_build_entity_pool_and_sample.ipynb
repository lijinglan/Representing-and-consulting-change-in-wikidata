# Build entity pool and 15,000 sample

This notebook creates the entity pool for the business domain and generates a random sample of 15,000 entities for later approach analysis.

The entity pool was retrieved from Wikidata Query Service using a SPARQL query. The full pool was exported as a CSV file. The 15,000 sample was then generated from this pool using Python with a fixed random seed.

## Step 1: Entity pool construction

The business entity pool was retrieved from Wikidata Query Service.

Retrieval date: 2026-06-09  
Endpoint: Wikidata Query Service  
Output file: `business_pool.csv`  
Number of entities: 765,042  

```sparql
SELECT DISTINCT ?item WHERE {
  ?item wdt:P31/wdt:P279* wd:Q4830453 .
}
```

This query retrieves entities whose instance-of class is business entity or one of its subclasses.

## Step 2: Random sample generation

The 15,000 business entity sample was generated from `business_pool.csv` using Python with a fixed random seed.

Random seed: 20  
Output file: `business_sample_15000.csv`

````python
import pandas as pd

df = pd.read_csv("business_pool.csv")

SEED = 20
sample_15000 = df.sample(n=15000, random_state=SEED).reset_index(drop=False)
sample_15000.to_csv("business_sample_15000.csv", index=False)

print('Successful.')
````
