```python
import pandas as pd

df = pd.read_csv("business-pool.csv")

SEED1 = 20
sample1_15000 = df.sample(n=15000, random_state=SEED1).reset_index(drop=False)
sample1_10000 = sample1_15000.iloc[:10000].reset_index(drop=True)  
sample1_5000  = sample1_15000.iloc[:5000].reset_index(drop=True)   

SEED2 = 21
sample2_10000 = df.sample(n=10000, random_state=SEED2).reset_index(drop=False)

SEED3 = 22
sample3_15000 = df.sample(n=15000, random_state=SEED3).reset_index(drop=False)

sample1_5000.to_csv("sample1_5000.csv",   index=False)
sample1_10000.to_csv("sample1_10000.csv", index=False)
sample1_15000.to_csv("sample1_15000.csv", index=False)
sample2_10000.to_csv("sample2_10000.csv", index=False)
sample3_15000.to_csv("sample3_15000.csv", index=False)

print('Successful.')
```

```python
# Check the duplication of the dataset to help determine if it is truly random

csv_5000  = pd.read_csv('sample1_5000.csv')
csv_10000  = pd.read_csv('sample2_10000.csv')
csv_15000  = pd.read_csv('sample3_15000.csv')

qids_5000  = set(csv_5000['item'].str.extract(r'(Q\d+)')[0].dropna())
qids_10000  = set(csv_10000['item'].str.extract(r'(Q\d+)')[0].dropna())
qids_15000 = set(csv_15000['item'].str.extract(r'(Q\d+)')[0].dropna())

overlap_all = qids_5000 & qids_10000 & qids_15000
overlap_5000_10000 = qids_5000 & qids_10000
overlap_10000_15000 = qids_10000 & qids_15000

print(f"Number of duplicate data in three datasets: {len(overlap_all)}")
print(f"The number of duplicate data in 5000 and 10000 datasets: {len(overlap_5000_10000)}")
print(f"The number of duplicate data in 10000 and 15000 datasets: {len(overlap_10000_15000)}")
```
