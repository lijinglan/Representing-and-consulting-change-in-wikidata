# Build entity pool and 15,000 sample

This notebook creates the entity pool for the politician domain and generates a random sample of 15,000 entities for later approach analysis.

The entity pool was retrieved from Wikidata Query Service using a SPARQL query. The full pool was exported as a CSV file. The 15,000 sample was then generated from this pool using Python with a fixed random seed.

## Step 1: Entity pool construction

The politician entity pool was retrieved from Wikidata Query Service.

Retrieval date: 2026-04-28  
Endpoint: Wikidata Query Service  
Output file: `politician_pool.csv`  
Number of entities: 929,221  

```sparql
SELECT DISTINCT ?item WHERE {
  ?item wdt:P31 wd:Q5 .
  ?item wdt:P106 wd:Q82955 .
}
```

This query retrieves entities that are instances of human and have politician as an occupation.

## Step 2: Random sample generation

The 15,000 business entity sample was generated from `politician_pool.csv` using Python with a fixed random seed.

Random seed: 20  
Output file: `politician_sample_15000.csv`

````python
import pandas as pd

df = pd.read_csv("politician_pool.csv")

SEED = 20
sample_15000 = df.sample(n=15000, random_state=SEED).reset_index(drop=False)
sample_15000.to_csv("politician_sample_15000.csv", index=False)

print('Successful.')
````
