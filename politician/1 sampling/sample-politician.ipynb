```python
import pandas as pd

df = pd.read_csv("politicianpool.csv")

SEED1 = 20
sample_15000 = df.sample(n=15000, random_state=SEED1).reset_index(drop=False)
sample_10000 = sample_15000.iloc[:10000].reset_index(drop=True)  
sample_5000  = sample_15000.iloc[:5000].reset_index(drop=True)   

sample_5000.to_csv("sample_5000.csv",   index=False)
sample_10000.to_csv("sample_10000.csv", index=False)
sample_15000.to_csv("sample_15000.csv", index=False)

print("Saved!")
```
