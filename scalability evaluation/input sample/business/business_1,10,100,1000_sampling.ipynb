# Create entity samples for scalability evaluation

This notebook creates nested entity samples for the scalability evaluation in the business domain.

Input file: `15000business_change_statements_final.xlsx`

Output files:  
`15000business_change_entities.xlsx`  
`business_scalability_sample_1.xlsx`  
`business_scalability_sample_10.xlsx`  
`business_scalability_sample_100.xlsx`  
`business_scalability_sample_1000.xlsx`

The entities are extracted from the final reviewed change-related statements. Therefore, each entity in the sampling frame has at least one identified change-related entity-property pair.

The scalability evaluation uses these entities as input sets for the query templates. Four nested samples are created, with sizes 1, 10, 100, and 1000. The smaller samples are subsets of the larger samples, so the results at different scales are more comparable.

These samples are used only for scalability testing. 

```python
import pandas as pd

doc = pd.read_excel("15000business_change_statements_final.xlsx")

entity_list = (
    doc[["QID"]]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
)

entity_list.to_excel(
    "15000business_change_entities.xlsx",
    index=False
)

print(f"Number of entities: {len(entity_list)}")
print("Saved to 15000business_change_entities.xlsx")
```

```python
df = pd.read_excel("15000business_change_entities.xlsx")

SEED1 = 21
business_scalability_sample_1000 = df.sample(n=1000, random_state=SEED1).reset_index(drop=False)
business_scalability_sample_100 = business_scalability_sample_1000.iloc[:100].reset_index(drop=True)  
business_scalability_sample_10 = business_scalability_sample_100.iloc[:10].reset_index(drop=True)  
business_scalability_sample_1  = business_scalability_sample_10.iloc[:1].reset_index(drop=True)   

business_scalability_sample_1000.to_excel("business_scalability_sample_1000.xlsx",   index=False)
business_scalability_sample_100.to_excel("business_scalability_sample_100.xlsx", index=False)
business_scalability_sample_10.to_excel("business_scalability_sample_10.xlsx", index=False)
business_scalability_sample_1.to_excel("business_scalability_sample_1.xlsx", index=False)
print('Successful.')
```
