```python
import pandas as pd

doc = pd.read_excel("manual15000business_change_statements.xlsx")

entity_list = (
    doc[["QID"]]
    .dropna()
    .drop_duplicates()
    .reset_index(drop=True)
)

entity_list.to_excel(
    "manual15000business_change_entities.xlsx",
    index=False
)

print(f"Number of entities: {len(entity_list)}")
print("Saved to manual15000business_change_entities.xlsx")
```
