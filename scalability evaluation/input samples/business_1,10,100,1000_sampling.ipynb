```python
import pandas as pd

df = pd.read_excel("manual15000business_change_entities.xlsx")

SEED1 = 21
sample_1000 = df.sample(n=1000, random_state=SEED1).reset_index(drop=False)
sample_100 = sample_1000.iloc[:100].reset_index(drop=True)  
sample_10 = sample_100.iloc[:10].reset_index(drop=True)  
sample_1  = sample_10.iloc[:1].reset_index(drop=True)   

sample_1000.to_excel("sample_1000.xlsx",   index=False)
sample_100.to_excel("sample_100.xlsx", index=False)
sample_10.to_excel("sample_10.xlsx", index=False)
sample_1.to_excel("sample_1.xlsx", index=False)
print('Successful.')
```
